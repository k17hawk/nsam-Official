```python
##Data analysis
##find all the missing values
```

```python
df_with_nan =[features for features in df.columns if df[features].isnull().sum()> 1]
```

```python
for feature in df_with_nan:
   print(feature,np.round(df[feature].isnull(),4),'%missing values')
```

```python
df_with_nan =[features for features in df.columns if df[features].isnull().sum()> 1]
for feature in df_with_nan:
    print(feature,np.round(df[feature].isnull().mean(),4),'% missing values')
```

```python
##see the relationship between dependent(sales price) and independent variable before droping the null values by ploting
```

```python
for feature in df_with_nan:
    data = df.copy()
    data[feature] = np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()  
```

```python
## 1 is our missing value. and from graph we see that the nan values dependent feature are also  related to independent feature.
```
